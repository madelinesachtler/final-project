# Data Cleaning
We will look to create two primary dataframes:
1. User-Friendly: A more complete DataFrame that can be interpreted directly.
2. Machine Learning: A DataFrame with only integer arguments, specifically catered to be used for machine learning

## Time of Day Field
We will be looking to create a new categorical variable (**Time of Day**) based on the **Game Completed Date**.  
This will look to categorize games on four times of the day that might provide some advantage for successful strategies that Imposters or Crewmates might use:
- **Night**: 12:00AM - 5:59AM  –  0, 1, 2, 3, 4, 5  
- **Morning**: 6:00AM - 11:59AM  –  6, 7, 8, 9, 10, 11  
- **Afternoon**: 12:00PM – 5:59PM  –  12, 13, 14, 15, 16, 17  
- **Evening**: 6:00PM-11:59PM  –  18, 19, 20, 21, 22, 23  

### Identifying the hour

Read **Game Completed Date** into a separate DataFrame to be split.  
We then redefine and select the columns of importance

```python
date = combined["Game Completed Date"]
cleaned_date = date.str.split(expand=True)
cleaned_date = cleaned_date.rename(columns={0:"Date", 2: "Time", 3:"Am/Pm"})
cleaned_date = cleaned_date[["Date", "Time", "Am/Pm"]]
cleaned_date
```

We will be converting the **Time** field into it's hour component.  
This is so we can convert it to it's Military Time counterpart later.

```python
cleaned_date['Time'] = pd.to_datetime(cleaned_date['Time'])
cleaned_date['Time'] = cleaned_date.Time.dt.hour
cleaned_date
```

### Convert hour to military time

There are 4 groups that we need to define in order to apply the appropriate manipulations.  
**AM Groups**  
12AM - Subtract 12 to convert to 0  
1PM-11PM - Leave as is  
**PM Groups**  
12PM - Leave as is  
1PM-11PM - Add 12 to convert to 13-23

Define the four groups as lists.

```python
am_no_change = np.arange(1, 12, 1).tolist()
am_change = [12]
pm_no_change = [12]
pm_change = np.arange(1, 12, 1).tolist()
```

Separate the data into separate DataFrames using 4 lists above and it's **Am/Pm** argument.

```python
am_no_change_df = cleaned_date.loc[(cleaned_date["Am/Pm"]=="am") & (cleaned_date["Time"].astype(int).isin(am_no_change))]
am_change_df = cleaned_date.loc[(cleaned_date["Am/Pm"]=="am") & (cleaned_date["Time"].astype(int).isin(am_change))]
pm_no_change_df = cleaned_date.loc[(cleaned_date["Am/Pm"]=="pm") & (cleaned_date["Time"].astype(int).isin(pm_no_change))]
pm_change_df = cleaned_date.loc[(cleaned_date["Am/Pm"]=="pm") & (cleaned_date["Time"].astype(int).isin(pm_change))]
```

Perform the necessary changes to get the military time for the two *change* groups.

```python
am_change_df["Time"] = am_change_df["Time"] - 12
pm_change_df["Time"] = pm_change_df["Time"] + 12
```

Rejoin the 4 DataFrame groups and reset the index so it reverts to it's original order.

```python
military_time_df = pd.concat([am_no_change_df, am_change_df, pm_no_change_df, pm_change_df])
military_time_df.sort_index(inplace = True)
military_time_df
```

### Time of Day Binning
Now that we have military time, we can easily bin these into the 4 **Time of Day** arguments.

Create the bin and create the name list for the four bins.

```python
bins = [0,6,12,18,24]

day_groups = ["Night", "Morning", "Afternoon", "Evening"]
```

Use *cut* create a new **Time of Day** column.

```python
military_time_df["Time of Day"] = pd.cut(military_time_df["Time"], bins, labels = day_groups, include_lowest = True)
military_time_df
```

## Convert ##m ##s to Seconds
For both **Game Length** and **Time to complete all tasks**, we need to convert this format into total seconds for analysis.

## Clean the data

Split the **Game Length** and **Time to complete all tasks** fields into separate columns in new DataFrames.

```python
cleaned_game_length = combined["Game Length"].str.split(expand=True)
cleaned_game_length = cleaned_game_length.rename(columns={0:"Minutes", 1: "Seconds"})

cleaned_task_length = combined["Time to complete all tasks"].str.split(expand=True)
cleaned_task_length = cleaned_task_length.rename(columns={0:"Minutes", 1: "Seconds"})
```

Remove all non-integer fields from each of the columns.

```python
cleaned_game_length['Minutes'] = cleaned_game_length['Minutes'].str.replace('m','').astype(int)

cleaned_game_length['Seconds'] = cleaned_game_length['Seconds'].str.replace('s','').astype(int)

cleaned_task_length['Minutes'] = cleaned_task_length['Minutes'].str.replace('-','0')
cleaned_task_length['Minutes'] = cleaned_task_length['Minutes'].str.replace('m','').astype(int)

i = 0
while i < len(cleaned_task_length['Seconds']):
    cleaned_task_length['Seconds'][i] = str('0' if cleaned_task_length['Seconds'][i] is None else cleaned_task_length['Seconds'][i])
    i += 1
cleaned_task_length['Seconds'] = cleaned_task_length['Seconds'].str.replace('s','').astype(int)
```

### Define the total number of seconds
Multiply **Minutes** by 60 and add to **Seconds** to convert **Seconds** into the total seconds.

```python
cleaned_game_length['Seconds'] = cleaned_game_length['Seconds'] + (cleaned_game_length['Minutes'] * 60)
cleaned_task_length['Seconds'] = cleaned_task_length['Seconds'] + (cleaned_task_length['Minutes'] * 60)
```

Pass back into the **combined** DataFrame to finalize the used-friendly DataFrame

```python
combined["Time_of_Day"] = military_time_df["Time of Day"]
combined
```

### Create Crewmate and Imposter DataFrames
For analytics and visualizations purposes, we split our data between **Team** lines.  
Additionally, we identify and drop columns that are wholly inapplicable to that team.

```python
crewmate_df = combined.loc[combined["Team"] == "Crewmate"].reset_index(drop=True)
crewmate_df.drop(columns = ["Imposter Kills"], inplace = True)

imposter_df = combined.loc[combined["Team"] == "Imposter"].reset_index(drop=True).reset_index(drop=True)
imposter_df.drop(columns = ["Task Completed", "All Tasks Completed", "Sabotages Fixed", "Time to complete all tasks", "Murdered", "Time to complete all tasks"], inplace = True)
```

## Binary Breakdown
For machine learning, we'll have to rewrite categorical fields into binary fields.

### Convert Time of Day
Take the newly created **Time of Day** field and split it into 4 binary columns.

```python
time_binary = pd.get_dummies(military_time_df, columns = ["Time of Day"])
time_binary
```

### Convert combined fields
Convert the **Team**, **Outcome**, **All Tasks Completed**, **Murdered**, and **Ejected** fields in **combined** into binary columns.

```python
combined_binary = pd.get_dummies(combined, columns = ["Team", "Outcome", "All Tasks Completed", "Murdered", "Ejected"])
combined_binary
```

Since we've just created many new columns, use *columns* to retrieve the exact field names.

```python
combined_binary.columns
```

## Binary Buildup
Now that we've created a binary breakdown for our categorical fields and converted ##m ##s to seconds, we'll construct a DataFrame with already appropriate fields from **combined**.  
While there exists non-integer values in some of the fields, these values exist along **Team** lines and will not be used for that team's analysis.

This finalizes the Machine Learning DataFrame.

```python
ml_df = pd.DataFrame({
    'User' : combined["User"],
    'Team' : combined_binary["Team_Crewmate"],
    'Outcome' : combined_binary["Outcome_Win"],
    'Game Length' : cleaned_game_length['Seconds'],
    'Task Completed' : combined["Task Completed"],
    'All Tasks Completed' : combined_binary["All Tasks Completed_Yes"],
    'Time to complete all tasks' : cleaned_task_length['Seconds'],
    'Imposter Kills' : combined["Imposter Kills"],
    'Sabotages Fixed' : combined["Sabotages Fixed"],
    'Murdered' : combined_binary["Murdered_Yes"],
    'Ejected' : combined_binary["Ejected_Yes"],
    'Night' : time_binary["Time of Day_Night"],
    'Morning' : time_binary["Time of Day_Morning"],
    'Afternoon' : time_binary["Time of Day_Afternoon"],
    'Evening' : time_binary["Time of Day_Evening"]
})
ml_df.head(20)
```

```python
ml_crewmate_df = ml_df.loc[ml_df["Team"] == 1].reset_index(drop=True)
ml_crewmate_df.drop(columns = ["Imposter Kills"], inplace = True)

ml_imposter_df = ml_df.loc[ml_df["Team"] == 0].reset_index(drop=True)
ml_imposter_df.drop(columns = ["Task Completed", "All Tasks Completed", "Sabotages Fixed", "Time to complete all tasks", "Murdered"], inplace = True)
```
